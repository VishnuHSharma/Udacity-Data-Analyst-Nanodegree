#### Importing the relevant libraries and initial analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
movies = pd.read_csv('tmdb-movies.csv')

movies.head()
```

```python
movies.info()
```

```python
movies.describe(include='all')
```

#### Based on initial analysis, we can see that there are a total of 10,866 samples in this dataset and 21 features. There are, however, numerous columns in which there are missing values. These majority of the missing values are homepage, tagline, keywords, and production companies. In this project, we are mainly interested in analyzing various quantative features such as budget and revenus and observing various trends over the years. Therfore, we can outright drop these columns. However, if we were interested in developing some sort of recommendation system, we would probably keep the keywords column.

```python
movies = movies.drop(['homepage', 'tagline', 'keywords', 'production_companies'], axis=1)
```

```python
movies.info()
```

#### The columns have been dropped.

### Are movie budgets increasing year over year?

```python
avg_budget_years = movies.groupby('release_year')['budget'].mean()
```

```python
def plot_asthetics(title, xlabel, ylabel):
    """  This is a custom function to avoid repeating the lines of code that
         write the titles, the axis of the plots
    """
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
avg_budget_years.plot(figsize=(10, 6))
plot_asthetics('Average Movie Budget(Non-Adjusted)', 'Year', 'Average Budget(Millions)')
```

#### As we can interpret from this chart, average movie budget has been increasing since movies first began, which again is understandable. It is also important to note that this dataset does not contain the more recent movies like Avengers. Also there are interesting peaks around the mid 2000s which could coincide with the release of Avatar.

### Are movies getting longer?

```python
average_runtime = movies.groupby('release_year')['runtime'].mean()
```

```python
average_runtime.plot(figsize=(10, 6))
plot_asthetics('Average Movie Runtime', 'Year', 'Average Runtime(Minutes)')
```

#### Again, this is something that could have been anticipated. Movie runtimes are actually getting shorter since the 1960s where they averaged around 2 hours and now closer to just over 90 minutes. Anectodotally, this makes sense as well.

#### Average Budget Adjusted plot over time.

```python
avg_budget_adj = movies.groupby('release_year')['budget_adj'].mean()
```

```python
avg_budget_adj.plot(figsize=(10, 6))
plot_asthetics('Average Movie Budget(Adjusted)', 'Year', 'Average Budget(Millions)')
```

### Who are some of the most prolific directors working in the industry?

```python
directors = movies['director'].value_counts().sort_values(ascending=False)
```

```python
top_ten = directors.head(10)
directors.head()
```

```python
top_ten.plot.barh(figsize=(10, 6))
plot_asthetics('Top Ten Movies Directors by Count', 'Movie Directed Count', 'Director')
```

### As we can see, within the constraints of this particular dataset, Woody Allen, Clint Eastwood, and Steven Spielberg have directed the most number of movies.
